```python
M = 1.75; #Madelung

a = [4.026*10**-10/2,2.02*10**-10,2.39*10**-10]; #LiF LiCl LiI
z = [9,17,53];
episilonr = [9.1,11.1,11.03]
z1 = 3;
e = 1.6e-19; #J
kspring = [];
M_a = 1.6605e-27 #kg
reduced_mass = []
Mli = 7*M_a
mass = [19*M_a,35.5*M_a,126.9*M_a]
w = []
rou = [2640,2070,4080]
r = []
for i in range(0,3):
    r0 = ((4*Mli+4*mass[i])/rou[i])**(1/3)
    r.append(r0/2)
```

Using $E = \frac{1}{4\pi \epsilon_{r} \epsilon_{0}}\frac{M \cdot e^{2}}{r}$, Taking derivative $k = \frac{\partial^2 E}{\partial r^2} = \frac{1}{4\pi \epsilon_{r} \epsilon_{0}} \frac{M \cdot e^{2}}{r^{3}}$

```python

for i in range(0,3):
    k = (M*(e**2)/((r[i])**3))*(1/(episilonr[i]*(1.112e-10)))
    kspring.append(k)
    reduce = (mass[i]*Mli)/(mass[i]+Mli)
    reduced_mass.append(reduce)
    omega = np.sqrt(kspring[i]/reduced_mass[i])
    w.append(omega/(10**12))
      
```
